```python
import os
import pandas as pd

# Path to training DICOM folders
dicom_root = "/content/MMGLF-colab-1/dataset/train"

# Path to tabular metadata CSV
csv_path = "/content/MMGLF-colab-1/dataset/tabular.csv"

# Output .txt path
output_txt = "/content/MMGLF-colab-1/train-oasis.txt"

# Load tabular data
df = pd.read_csv(csv_path)

# Create train-oasis.txt
with open(output_txt, 'w') as f:
    for subject in os.listdir(dicom_root):
        subject_path = os.path.join(dicom_root, subject)
        if not os.path.isdir(subject_path):
            continue
        row = df[df["PTID"] == subject]
        if not row.empty:
            cdr = float(row.iloc[0]["CDR"])
            label = 1 if cdr >= 0.5 else 0
            f.write(f"{subject_path} {label}\n")

print("‚úÖ Created:", output_txt)
```

```python
# For val-oasis.txt
val_dicom_root = "/content/MMGLF-colab-1/dataset/val"
val_output_txt = "/content/MMGLF-colab-1/val-oasis.txt"

with open(val_output_txt, 'w') as f:
    for subject in os.listdir(val_dicom_root):
        subject_path = os.path.join(val_dicom_root, subject)
        if not os.path.isdir(subject_path):
            continue
        row = df[df["PTID"] == subject]
        if not row.empty:
            cdr = float(row.iloc[0]["CDR"])
            label = 1 if cdr >= 0.5 else 0
            f.write(f"{subject_path} {label}\n")

print("‚úÖ Created:", val_output_txt)
```

```python
import os
import pandas as pd
from glob import glob

# Define paths
base_path = "/content/MMGLF-colab-1/dataset"
csv_path = os.path.join(base_path, "tabular.csv")
train_dir = os.path.join(base_path, "train")
val_dir = os.path.join(base_path, "val")
train_txt = "/content/MMGLF-colab-1/train-oasis.txt"
val_txt = "/content/MMGLF-colab-1/val-oasis.txt"

# Load tabular CSV
df = pd.read_csv(csv_path)

# Fix: Use 'ID' instead of 'PTID'
df['ID'] = df['ID'].astype(str).str.strip()

# Helper function to create txt file
def create_txt(data_dir, output_txt, df):
    with open(output_txt, 'w') as f:
        nii_files = glob(os.path.join(data_dir, "*.nii*"))
        for file_path in nii_files:
            filename = os.path.basename(file_path)

            # Try to extract matching ID (like OAS1_0111_MR1) from filename
            for id_val in df["ID"]:
                if id_val in filename:
                    cdr = float(df[df["ID"] == id_val].iloc[0]["CDR"])
                    label = 1 if cdr >= 0.5 else 0
                    f.write(f"{file_path} {label}\n")
                    break

create_txt(train_dir, train_txt, df)
create_txt(val_dir, val_txt, df)

print("‚úÖ Created:")
print("   -", train_txt)
print("   -", val_txt)
```
