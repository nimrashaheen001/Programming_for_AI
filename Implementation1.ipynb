```python
import os
import shutil
import random

def balance_dataset_by_undersampling(source_dir, output_dir, image_extensions=('.dcm',)):
    """
    Balances a dataset by undersampling all classes to the size of the smallest class.

    - source_dir: original dataset directory where each class is a subfolder
    - output_dir: directory to save the balanced dataset
    - image_extensions: tuple of file extensions to include (default: DICOM files only)
    """
    os.makedirs(output_dir, exist_ok=True)

    # Find class folders (MCI, AD, NC)
    class_dirs = [d for d in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, d))]

    # Count images in all subfolders (like MCI/21f/, MCI/15f/, etc.)
    class_counts = {}
    image_paths_per_class = {}

    for class_dir in class_dirs:
        class_path = os.path.join(source_dir, class_dir)
        all_images = []

        # Traverse subfolders (like 21f, 15f, etc.)
        for root, _, files in os.walk(class_path):
            for file in files:
                if file.lower().endswith(image_extensions):
                    all_images.append(os.path.join(root, file))

        class_counts[class_dir] = len(all_images)
        image_paths_per_class[class_dir] = all_images

    min_count = min(class_counts.values())
    print(f"Balancing to {min_count} images per class.")

    for class_dir in class_dirs:
        dest_class_dir = os.path.join(output_dir, class_dir)
        os.makedirs(dest_class_dir, exist_ok=True)

        selected_images = random.sample(image_paths_per_class[class_dir], min_count)

        for i, img_path in enumerate(selected_images):
            # Use flat filenames to avoid overwriting
            dest_path = os.path.join(dest_class_dir, f"{class_dir}_{i}.dcm")
            shutil.copy(img_path, dest_path)

    print(f"Balanced dataset saved to: {output_dir}")
source_path = '/content/unzipped_dataset/Alzheimers disease MRI images'
balanced_path = 'balanced_dataset_dcm'

balance_dataset_by_undersampling(source_path, balanced_path)
```
